```python
import pandas as pd
import numpy as np

dataframe = pd.read_csv("../dados/CC GENERAL.csv")
dataframe.head()
```

```python
np.unique(dataframe['TENURE'])
```

```python
# não agregam valor à classificação ou clasterização. 
#     CUST_ID: ID do cliente, não ajuda a classificá-lo.
#     TENURE : tempo para renovação do cartão de cŕedito.
dataframe.drop(columns=["CUST_ID", "TENURE"], inplace=True)   
```

```python
dataframe.head(12)
```

```python
missing = dataframe.isna().sum()
print(missing)
```

```python
# Completar valores c/ mediana: não é o valor correto, mas aproximado.
dataframe.fillna(dataframe.median(), inplace=True)
missing = dataframe.isna().sum()
print(missing)
```

```python
# Há atributos que não estão no mesmo intervalo. 
#    Isso pode prejudicar o algoritmo.
#    Normalização: colocaremos tudo entre 0 e 1. 
#    Notação científica: 
#      1.34239194e-01 = 1.34239194 × 10^(⁻1) = 0.134239194 (POSITIVO)
from sklearn.preprocessing import Normalizer
values = Normalizer().fit_transform(dataframe.values)
print(values)
```

```python
# Executando o clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, n_init=10, max_iter=300)
y_pred = kmeans.fit_predict(values)
```

```python
# Os clústeres formados
print('Clústeres formados:', *(np.unique(y_pred)), sep='\n -> ')
```
